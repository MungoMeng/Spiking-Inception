```python
import numpy as np
import os.path
import scipy 
import brian2 as b
from brian2 import *
import time
from math import *

from Functions import *

#set parameters
np.random.seed(0)
prefs.codegen.target = 'auto'
prefs.codegen.cpp.extra_compile_args_gcc = ['-march=native']

#---------------------------------------------------Build network----------------------------------------------------------
Learning = True

#the amount of stacked module
module_num = 3
print 'Num of stacked module:', module_num

print 'The settings of modules are as follow:'
print '---------------------------------------------'

#the amount of input
n_input = 784
module_input = np.array([1,2,4])
print 'Num of input:', module_input * n_input

#connection parameters
kernel_type_num = 3
kernel_size_each_type = [28, 24, 16]
stride_each_type = [1, 4, 6]
feature_map_size_each_type = [int(((sqrt(n_input) - kernel_size_each_type[i]) / stride_each_type[i] + 1)**2) for i in range(kernel_type_num)]
print 'Num of kernel type:', kernel_type_num
print 'Kernel size and Stride of each kernel type:', kernel_size_each_type, stride_each_type
print 'Feature map size of each kernel type:', feature_map_size_each_type

feature_map_num = [112,224,448,448]
kernel_num_each_type = [[1, 1, 1],[1, 1, 1],[1, 1, 1]]
kernel_num = np.sum(kernel_num_each_type,axis=1)
print 'Feature map num of each module:', feature_map_num
print 'Kernel num of each kernel type:', kernel_num_each_type
print 'Num of kernel:', kernel_num

#the amount of neurons
neuron_num_each_kernel = []
feature_map_size_each_kernel = []
kernel_size_each_kernel = []
stride_each_kernel = []
for module in range(module_num):
    neuron_num_temp = []
    feature_map_size_temp = []
    kernel_size_temp = []
    stride_temp = []
    for kernel_type in range(kernel_type_num):
        for kernel in range(kernel_num_each_type[module][kernel_type]):
            neuron_num_temp.append(feature_map_num[module] * feature_map_size_each_type[kernel_type])
            feature_map_size_temp.append(feature_map_size_each_type[kernel_type])
            kernel_size_temp.append(kernel_size_each_type[kernel_type])
            stride_temp.append(stride_each_type[kernel_type])
    neuron_num_each_kernel.append(neuron_num_temp)
    feature_map_size_each_kernel.append(feature_map_size_temp)
    kernel_size_each_kernel.append(kernel_size_temp)
    stride_each_kernel.append(stride_temp)

neuron_num = []
for module in range(module_num):
    neuron_num.append(np.sum(neuron_num_each_kernel[module]))
print 'Neurons num of each kernel:', neuron_num_each_kernel
print 'Num of Neurons:', neuron_num

pool_size = []
pool_padding = []
for module in range(module_num-1):
    pool_size.append(int(ceil(float(neuron_num[module])/float(module_input[module+1]*n_input))))
    pool_padding.append(int(pool_size[module] * module_input[module+1]*n_input - neuron_num[module]))
print 'Pool size of connection between modules', pool_size
print 'Padding of connection between modules', pool_padding
print '---------------------------------------------'
    
#neuron parameters
v_rest_e = -65. * b.mV
v_reset_e = -65. * b.mV
v_thresh_e = -52. * b.mV
refrac_e = 5. * b.ms

#synapses parameters
Delay = 10*b.ms
tc_pre_ee = 20*b.ms
tc_post_1_ee = 20*b.ms
tc_post_2_ee = 40*b.ms
nu_ee_pre =  0.0001
nu_ee_post = 0.01
wmax_ee = 1.0
wmin_ee = 1e-7
ihn = 24
norm = 78.4
    
if Learning == False:
    scr_e = 'v = v_reset_e'
else:
    tc_theta = 1e7 * b.ms
    theta_plus_e = 0.05 * b.mV
    scr_e = 'v = v_reset_e; theta += theta_plus_e'
offset = 20.0*b.mV
thresh_e = 'v>(theta - offset + ' + str(v_thresh_e/b.mV) + '*mV' + ')'

#equation of excitatory neurons
neuron_eqs_e = '''
            dv/dt = ((v_rest_e - v) + (I_synE+I_synI) / nS) / (100*ms)  : volt
            I_synE = ge * nS * (        -v)                             : amp
            I_synI = gi * nS * (-100.*mV-v)                             : amp
            dge/dt = -ge/(1.0*ms)                                       : 1
            dgi/dt = -gi/(2.0*ms)                                       : 1
            '''
if Learning == False:
    neuron_eqs_e += '\n  theta      :volt'
else:
    neuron_eqs_e += '\n  dtheta/dt = -theta / (tc_theta)  : volt'

#equation of STDP
eqs_stdp_ee = '''
                    w                                      : 1
                    post2before                            : 1
                    dpre/dt    = -pre/(tc_pre_ee)          : 1 (event-driven)
                    dpost1/dt  = -post1/(tc_post_1_ee)     : 1 (event-driven)
                    dpost2/dt  = -post2/(tc_post_2_ee)     : 1 (event-driven)
                '''
eqs_stdp_pre_ee = 'ge+=w; pre = 1.; w = (w>0)*clip(w - nu_ee_pre * post1 , wmin_ee, wmax_ee)'
eqs_stdp_post_ee = 'post2before = post2; w = (w>0)*clip(w + nu_ee_post * pre * post2before, wmin_ee, wmax_ee); post1 = 1.; post2 = 1.'
    
#create empty dict
neuron_groups = {}
connections = {}
spike_counters = {}
net = {}
    
#create neuron group
for i in range(module_num):
    neuron_groups['X'+str(i+1)] = b.PoissonGroup(module_input[i] * n_input, 0*b.hertz)
    neuron_groups['A'+str(i+1)] = b.NeuronGroup(neuron_num[i], neuron_eqs_e, method='euler', threshold=thresh_e, refractory=refrac_e, reset= scr_e)
    neuron_groups['A'+str(i+1)].v = v_rest_e - 40. * b.mV
    neuron_groups['A'+str(i+1)].theta = np.ones((neuron_num[i])) * 20.0*b.mV
    
#create connections AA
for i in range(module_num):
    start = time.time()
    weightMatrix = np.zeros((neuron_num[i], neuron_num[i]))
    mark = 0
    for kernel in range(kernel_num[i]):
        feature_map_size = feature_map_size_each_kernel[i][kernel]
        for src in range(mark, mark+neuron_num_each_kernel[i][kernel]):
            S = src - mark
            src_z = int(S/feature_map_size)
            src_y = int((S - src_z*feature_map_size) / sqrt(feature_map_size))
            src_x = int(S - src_z*feature_map_size - src_y*sqrt(feature_map_size))
            for tar in range(mark, mark+neuron_num_each_kernel[i][kernel]):
                T = tar - mark
                tar_z = int(T / feature_map_size)
                tar_y = int((T - tar_z*feature_map_size) / sqrt(feature_map_size))
                tar_x = int(T - tar_z*feature_map_size - tar_y*sqrt(feature_map_size))
                if src_x == tar_x and src_y == tar_y and src_z != tar_z:
                    weightMatrix[src,tar] = ihn
        mark += neuron_num_each_kernel[i][kernel]
    weightMatrix = weightMatrix.reshape((neuron_num[i]*neuron_num[i]))
    connections['A'+str(i+1)+'A'+str(i+1)] = b.Synapses(neuron_groups['A'+str(i+1)], neuron_groups['A'+str(i+1)], 'w:1',on_pre='gi+=w')
    connections['A'+str(i+1)+'A'+str(i+1)].connect()
    connections['A'+str(i+1)+'A'+str(i+1)].w = weightMatrix
    end = time.time()
    print 'time needed to create connection A'+str(i+1)+'A'+str(i+1)+':', end - start

#create connections XA
for i in range(module_num):
    start = time.time()
    weightMatrix = np.zeros((module_input[i]*n_input, neuron_num[i]))
    if Learning:
        mark = 0
        for kernel in range(kernel_num[i]):
            feature_map_size = feature_map_size_each_kernel[i][kernel]
            kernel_size = kernel_size_each_kernel[i][kernel]
            stride = stride_each_kernel[i][kernel]
            for src in range(module_input[i]*n_input):
                src_z = int(src / n_input)
                src_y = int((src - src_z*n_input) / sqrt(n_input))
                src_x = int(src - src_z*n_input - src_y*sqrt(n_input))
                for tar in range(mark, mark+neuron_num_each_kernel[i][kernel]):
                    T = tar - mark
                    tar_z = int(T / feature_map_size)
                    tar_y = int((T - tar_z*feature_map_size) / sqrt(feature_map_size))
                    tar_x = int(T - tar_z*feature_map_size - tar_y*sqrt(feature_map_size))
                    if src_x >= tar_x*stride and src_x < tar_x*stride+kernel_size and src_y >= tar_y*stride and src_y < tar_y*stride+kernel_size:
                        weightMatrix[src,tar] = 0.3*rand()+wmin_ee
            mark += neuron_num_each_kernel[i][kernel]
    weightMatrix = weightMatrix.reshape((module_input[i]*n_input*neuron_num[i]))

    if Learning:
        connections['X'+str(i+1)+'A'+str(i+1)] = b.Synapses(neuron_groups['X'+str(i+1)], neuron_groups['A'+str(i+1)], eqs_stdp_ee, on_pre=eqs_stdp_pre_ee, on_post=eqs_stdp_post_ee)
    else:
        connections['X'+str(i+1)+'A'+str(i+1)] = b.Synapses(neuron_groups['X'+str(i+1)], neuron_groups['A'+str(i+1)], 'w : 1', on_pre='ge+=w')
    connections['X'+str(i+1)+'A'+str(i+1)].connect()
    connections['X'+str(i+1)+'A'+str(i+1)].w = weightMatrix
    connections['X'+str(i+1)+'A'+str(i+1)].delay = 'rand()*'+ str(Delay/b.ms) +'*ms'
    end = time.time()
    print 'time needed to create connection X'+str(i+1)+'A'+str(i+1)+':', end - start

#create monitors
for i in range(module_num):
    spike_counters['A'+str(i+1)] = b.SpikeMonitor(neuron_groups['A'+str(i+1)], record=False)

#create networks
for i in range(module_num):
    net['M'+str(i+1)] = Network(neuron_groups['A'+str(i+1)],neuron_groups['X'+str(i+1)],connections['X'+str(i+1)+'A'+str(i+1)],connections['A'+str(i+1)+'A'+str(i+1)],spike_counters['A'+str(i+1)])
#-----------------------------------------------------------------------------------------------------------------------
```

```python
# load MNIST
start = time.time()
training = get_labeled_data('./training')
end = time.time()
print 'time needed to load training set:', end - start

#specify the location
save_path = './weights/'
load_path = './weights/'

#the time-window of simulation
single_example_time =   0.35 * b.second
resting_time = 0.15 * b.second

#the the interval of process data and show information
progress_interval = 1
validate_interval = 500
save_interval = 100

#number of samples for training
n_train = 60000
train_begin = 0

#load trained weight to continue
if train_begin:
    for i in range(module_num):
        connections['X'+str(i+1)+'A'+str(i+1)].w = np.load(load_path + 'X' + str(i+1) + 'A' + str(i+1) + '_' + str(train_begin) + '.npy')
        neuron_groups['A'+str(i+1)].theta = np.load(load_path + 'theta_A' + str(i+1) + '_' + str(train_begin) + '.npy') *b.volt
```

```python
#the intensity of rate coding
intensity_step = 0.25
start_intensity = [0.25,1.0,1.0]

#the threshold of retrain
retrain_gate = []
for i in range(module_num):
    retrain_gate.append(np.sum([5*feature_map_size_each_kernel[i][kernel] for kernel in range(kernel_num[i])]))

# run the simulation and set inputs
previous_spike_count = {}
current_spike_count = {}
assignments = {}
result_monitor = {}
results_proportion = {}
accuracy = {}
for i in range(module_num):
    previous_spike_count['A'+str(i+1)] = np.zeros(neuron_num[i])
    current_spike_count['A'+str(i+1)] = np.zeros(neuron_num[i])
    assignments['A'+str(i+1)] = np.zeros(neuron_num[i])
    result_monitor['A'+str(i+1)] = np.zeros((validate_interval,neuron_num[i]))
    results_proportion['A'+str(i+1)] = np.zeros((10, validate_interval))
    accuracy['A'+str(i+1)] = []
input_numbers = np.zeros(validate_interval)

for i in range(module_num):
    neuron_groups['X'+str(i+1)].rates = 0*b.hertz
    net['M'+str(i+1)].run(0*b.second)

start = time.time()
j = train_begin
while j < n_train:
    m = 0
    input_intensity = start_intensity[m]
    Rates = training['x'][j%60000,:,:].reshape((n_input)) * input_intensity
    while m < module_num:
        neuron_groups['X'+str(m+1)].rates = Rates*b.hertz
        connections['X'+str(m+1)+'A'+str(m+1)] = normalize_weights(connections['X'+str(m+1)+'A'+str(m+1)],norm*module_input[m])

        net['M'+str(m+1)].run(single_example_time)
    
        current_spike_count['A'+str(m+1)] = np.asarray(spike_counters['A'+str(m+1)].count[:])- previous_spike_count['A'+str(m+1)]
        previous_spike_count['A'+str(m+1)] = np.copy(spike_counters['A'+str(m+1)].count[:])
    
        #if current_spike_count is not enough, increase the input_intensity and simulat this example again
        spike_num = np.sum(current_spike_count['A'+str(m+1)])
        print spike_num,' for ', m

        if spike_num < retrain_gate[m]:
            Rates = Rates / input_intensity
            input_intensity += intensity_step
            Rates = Rates * input_intensity
            neuron_groups['X'+str(m+1)].rates = 0*b.hertz
            net['M'+str(m+1)].run(resting_time)
        else:
            result_monitor['A'+str(m+1)][j%validate_interval,:] = current_spike_count['A'+str(m+1)]
            input_numbers[j%validate_interval] = training['y'][j%60000][0]

            neuron_groups['X'+str(m+1)].rates = 0*b.hertz
            net['M'+str(m+1)].run(resting_time)
            
            #use the output for the input of the next module
            m += 1
            if m < module_num:
                input_intensity = start_intensity[m]
                spike_count_padding = np.hstack((current_spike_count['A'+str(m)],np.zeros(pool_padding[m-1])))
                Rates = np.array([np.sum(spike_count_padding[i*pool_size[m-1]:(i+1)*pool_size[m-1]]) for i in range(module_input[m]*n_input)])
                Rates = Rates / np.mean(Rates[Rates!=0]) * 255
                Rates = Rates * input_intensity

    j += 1
    if j % progress_interval == 0:
        print 'Progress: ', j, '/', n_train, '(', time.time() - start, 'seconds)'
        start = time.time()
    if j % validate_interval == 0:
        for i in range(module_num):
            assignments['A'+str(i+1)] = get_new_assignments(result_monitor['A'+str(i+1)][:], input_numbers[:])
            test_results = np.zeros((10, validate_interval))
            for k in range(validate_interval):
                results_proportion['A'+str(i+1)][:,k] = get_recognized_number_proportion(assignments['A'+str(i+1)], result_monitor['A'+str(i+1)][k,:])
                test_results[:,k] = np.argsort(results_proportion['A'+str(i+1)][:,k])[::-1]
            difference = test_results[0,:] - input_numbers[:]
            correct = len(np.where(difference == 0)[0])
            accuracy['A'+str(i+1)].append(correct/float(validate_interval) * 100)
            print 'Validate accuracy for A'+str(i+1)+': ', accuracy['A'+str(i+1)][-1], '(last)', np.max(accuracy['A'+str(i+1)]), '(best)'
            
    if j % save_interval == 0:
        for i in range(module_num):
            np.save(save_path + 'X' + str(i+1) + 'A' + str(i+1) + '_' + str(j), connections['X'+str(i+1)+'A'+str(i+1)].w)
            np.save(save_path + 'theta_A' + str(i+1) + '_' + str(j), neuron_groups['A'+str(i+1)].theta)
```
